```python

import os
import pyautogui
import shutil
import time

from watchdog.observers import Observer
from watchdog.events import FileSystemEventHandler

# DEFININDO PASTAS
pasta_origem = "C:\\Users\\ti-me\\OneDrive\\Documentos\\jupyter\\auxiliares\\pasta_origem"
pasta_destino = "C:\\Users\\ti-me\\OneDrive\\Documentos\\jupyter\\auxiliares\\pasta_destino"
running = False

def illustrator() :
    illustrator_path = "C:\\Program Files\\Adobe\\Adobe Illustrator 2022\\Support Files\\Contents\\Windows\\Illustrator"
    template_path = "X:\Impressao_IML\Templates_VDT\Template_VDT_400ml_ok.ai"
    planilha = "X:\\Impressao_IML\\Lote 495\\Planilha_lote_495_400ml.csv"
    pastaArtes = "X:\\Impressao_IML\\Lote 495\\Artes PDF"
    pastaVDT = "X:\\Impressao_IML\\Lote 495"

    os.startfile(illustrator_path)
    time.sleep(3)
    #
    # abrindo o template
    pyautogui.moveTo(102, 11)
    time.sleep(1)
    pyautogui.click(button='left')
    time.sleep(2)
    pyautogui.moveTo(167, 77)
    time.sleep(1)
    pyautogui.click(button='left')
    time.sleep(3)
    pyautogui.write(template_path)
    time.sleep(1)
    pyautogui.press('enter')
    time.sleep(3)
    #
    #VDP
    pyautogui.moveTo(1731, 284)
    time.sleep(2)
    pyautogui.click(button='left')
    time.sleep(2)
    pyautogui.moveTo(1648, 292)
    time.sleep(2)
    pyautogui.click(button='left')
    time.sleep(2)
    pyautogui.moveTo(987, 354)
    time.sleep(2)
    pyautogui.click(button='left')
    time.sleep(2)
    pyautogui.moveTo(1160, 459)
    time.sleep(2)
    pyautogui.click(button='left')
    time.sleep(2)
    pyautogui.write(planilha)
    time.sleep(2)
    pyautogui.press('enter')
    time.sleep(3)
    pyautogui.moveTo(1063, 640)
    time.sleep(2)
    pyautogui.click(button='left')
    time.sleep(2)
    pyautogui.moveTo(1085, 759)
    time.sleep(2)
    pyautogui.click(button='left')
    time.sleep(2)
    pyautogui.moveTo(1585, 355)
    time.sleep(2)
    pyautogui.click(button='left')
    pyautogui.click(button='left')
    time.sleep(2)
    pyautogui.moveTo(1648, 628)
    time.sleep(2)
    pyautogui.click(button='left')
    time.sleep(2)
    pyautogui.moveTo(1589, 606)
    time.sleep(2)
    pyautogui.click(button='left')
    time.sleep(2)
    pyautogui.write(pastaArtes)
    time.sleep(3)
    pyautogui.moveTo(965, 560)
    time.sleep(2)
    pyautogui.click(button='left')
    time.sleep(2)
    pyautogui.moveTo(1465, 283)
    time.sleep(2)
    pyautogui.click(button='left')
    time.sleep(2)
    pyautogui.moveTo(1731, 284)
    time.sleep(2)
    pyautogui.click(button='left')
    time.sleep(2)
    pyautogui.moveTo(1562, 435)
    time.sleep(2)
    pyautogui.click(button='left')
    pyautogui.moveTo(690, 77)
    time.sleep(2)
    pyautogui.click(button='left')
    time.sleep(2)
    pyautogui.write(pastaVDT)
    time.sleep(1)
    pyautogui.press('enter')
    time.sleep(2)
    pyautogui.moveTo(965, 560)
    time.sleep(2)
    pyautogui.click(button='left')

illustrator()

def photoshop() :
    photoshop_path = "C:\\ProgramData\\Microsoft\\Windows\\Start Menu\\Programs\\Adobe Photoshop 2024"
    template_path = "X:\Impressao_IML\\templatesPhotoshop\Template_Polytype\Template_300ml_polytype.psd"
    calco_branco_path = "X:\Impressao_IML\Lote 494\Calco Branco.txt"
    psd_files_path = "X:\Impressao_IML\Lote 494\Arquivos PSD"
    pdf_arts_path = "X:\Impressao_IML\Lote 494\Artes PDF"
    running = True
    
    # abrindo o photoshop
    os.startfile(photoshop_path)
    time.sleep(3)
    #
    # abrindo o template
    pyautogui.hotkey('ctrl', 'o')
    time.sleep(1)
    pyautogui.write(template_path)
    time.sleep(1)
    pyautogui.press('enter')
    time.sleep(3)
    #
    # abrindo o calço branco
    pyautogui.hotkey('alt', 'i')
    time.sleep(1)
    pyautogui.press('v')
    time.sleep(1)
    pyautogui.press('c')
    time.sleep(1)
    pyautogui.hotkey('alt', 'i')
    time.sleep(1)
    pyautogui.hotkey('alt', 's')
    time.sleep(1)
    pyautogui.write(calco_branco_path)
    time.sleep(1)
    pyautogui.press('enter')
    time.sleep(1)
    pyautogui.press('enter')
    pyautogui.moveTo(1554, 195)
    time.sleep(1)
    pyautogui.click(button='left')
    time.sleep(5)
    # 
    # exportando arquivos PSD
    pyautogui.hotkey('alt', 'a')
    time.sleep(1)
    pyautogui.press('x')
    time.sleep(1)
    pyautogui.moveTo(413, 588)
    time.sleep(2)
    pyautogui.click(button='left')
    time.sleep(1)
    pyautogui.moveTo(944, 335)
    pyautogui.click(button='left')
    time.sleep(1)
    pyautogui.write(psd_files_path)
    time.sleep(1)
    pyautogui.press('enter')
    time.sleep(1)
    pyautogui.press('enter')
    time.sleep(1)
    pyautogui.press('enter')
    
    time.sleep(40) # tempo de geração das imagens
    
    #fechar projeto
    # time.sleep(3)
    # pyautogui.moveTo(345, 77)
    # pyautogui.click(button='left')
    # time.sleep(1)
    # pyautogui.press('right')
    # time.sleep(1)
    # pyautogui.press('enter')
    #
    
    # Automatizando lote
    time.sleep(3)
    pyautogui.hotkey('alt', 'a')
    time.sleep(1)
    pyautogui.press('z')
    time.sleep(1)
    pyautogui.press('l')
    time.sleep(1)
    pyautogui.moveTo(898, 243)
    pyautogui.click(button='left')
    time.sleep(1)
    pyautogui.moveTo(880, 281)
    pyautogui.click(button='left')
    time.sleep(1)
    pyautogui.moveTo(898, 273)
    pyautogui.click(button='left')
    time.sleep(1)
    pyautogui.moveTo(891, 309)
    pyautogui.click(button='left')
    time.sleep(1)
    pyautogui.moveTo(812, 340)
    pyautogui.click(button='left')
    time.sleep(1)
    pyautogui.write(psd_files_path)
    time.sleep(1)
    pyautogui.press('enter')
    time.sleep(1)
    pyautogui.moveTo(1214, 238)
    pyautogui.click(button='left')
    time.sleep(1)
    pyautogui.write(pdf_arts_path)
    time.sleep(1)
    pyautogui.press('enter')
    time.sleep(1)
    pyautogui.moveTo(1740, 217)
    pyautogui.click(button='left')
    
    running = False
    
photoshop()

def verify_and_move_initial_files() :
    arquivos = os.listdir(pasta_origem)

    if arquivos and not running:
        print(arquivos)
        for arquivo in arquivos :
            caminho_origem = os.path.join(pasta_origem, arquivo)
            caminho_final = os.path.join(pasta_destino, arquivo)
            
            shutil.move(caminho_origem, caminho_final);
            
            print(f"{arquivo} movido para {pasta_destino}")
    else :
        print("Nenhum arquivo encontrado")
        

verify_and_move_initial_files() # Verifica todos os arquivos que estão na pasta e move eles 

# CLASSE PARA MANUSEAR OS EVENTOS DOS ARQUIVOS
class MoveFilesHandler(FileSystemEventHandler) :
    def on_created(self, event) :
        if not event.is_directory:
            arquivo = os.path.basename(event.src_path)
            caminho_destino = os.path.join(pasta_destino, arquivo)

            shutil.move(event.src_path, caminho_destino)
            print(f"{arquivo} movido para {pasta_destino}")

# Configura o observador
observador = Observer()
evento_handler = MoveFilesHandler()
observador.schedule(evento_handler, pasta_origem, recursive=False)

# Inicia a observação da pasta
observador.start()
print("Monitorando a pasta para novos arquivos...")

try : 
    while True:
        time.sleep(10)
except KeyboardInterrupt :
    observador.stop()
observador.join() # Finaliza qualquer operação pendente do observador
```
